```python
df_4c = a.query("country in ('United States','Finland','India','Venezuela')")
```

```python
(
so.Plot(df_4c, x="year", y="life_expect", color="country")
.add(so.Line())
)
```

```python
(
    so.Plot(df_4c,x="life_expect",y="country").add(so.Bar(color="g"))
)
```

```python
so.Plot(df_4c,x="pop",y="year").add(so.Bar(color="g"))
```

```python
df_4c.dtypes
```

```python
df_4c['year'] = df_4c['year'].astype('category')
```

```python
df_4c.dtypes
```

```python
so.Plot(df_4c,x="pop",y="year").add(so.Bar(color="g"))
```

```python
(
    so.Plot(data =df_4c,x = 'country',y = 'life_expect').add(so.Bar(), so.Agg('mean'))
)
```

```python
(
    so.Plot(data =df_4c,x = 'country',y = 'life_expect').add(so.Bar(), so.Agg('median'))
)
```

```python
(
    so.Plot(data =df_4c,x = 'country',y = 'life_expect').add(so.Bar(), so.Agg('sum'))
)
```

```python
(
    so.Plot(data =df_4c,x = 'country',y = 'life_expect').add(so.Bar(), so.Agg('min'))
)
```

```python
(
    so.Plot(data =df_4c,x = 'country',y = 'life_expect').add(so.Bar(), so.Agg('max'))
)
```

```python
g = sns.catplot(
data=df_4c, kind="bar",
x="country", y="fertility", hue="country",
errorbar="sd", palette="dark", alpha=.6, height=6
)
```

```python
print (type (g))
```

```python
print(g.legend)
```

```python
type (g.legend) == type (g)
```

```python
g.despine(left=True)
g.set_axis_labels("Country name", "fertility")
g.legend.set_title("4 countries Vs fertility")
```
